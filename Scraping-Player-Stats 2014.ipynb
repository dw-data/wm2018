```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import re
import os
import requests
import dateutil
from datetime import datetime

pd.set_option("display.max_rows", 1000)
```

```python
start_page_url = 'http://www.fifa.com/worldcup/archive/brazil2014/statistics/players/distance.html'
```

```python
requested_page = Request(start_page_url, headers={'User-Agent': 'Chrome/64.0'})
webpage = urlopen(requested_page).read()
overview_page = BeautifulSoup(webpage, "html.parser")
```

```python
names_list = []
for td in overview_page.findAll('td', {'class':'tbl-playername'}):
    names_list.append(td.text)
    
matches_list = []
for td in overview_page.findAll('td', {'class':'tbl-mp'}):
    matches_list.append(td.text)

minutes_list = []
for td in overview_page.findAll('td', {'class':'tbl-minp'}):
    minutes_list.append(td.text)
```

```python
player_stats = pd.DataFrame()
player_stats['name'] = [element for element in names_list]
player_stats['number_matches'] = [element for element in matches_list]
player_stats['minutes_played'] = [element for element in minutes_list]
player_stats.head()
```

```python
def keep_uppercase(mixed_string):
    uppercase_string = re.sub('(\D*\s)([A-Z]+$)',r'\2',mixed_string)
    return uppercase_string

player_stats['lastname'] = player_stats['name'].apply(keep_uppercase)
player_stats
```

```python
player_stats.to_csv('data/output/player_stats.csv', encoding = 'utf-8')
```
