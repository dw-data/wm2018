```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 1000)

plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
```

```python
df = pd.read_csv('data/output/Data-World-Cup-2018.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
played = df[df['Minutes_played']>0]
played.head()
```

## List of national teams

```python
played['Nationalmannschaft'].value_counts()
```

## List of Leagues

```python
played['League_ID'] = played['Club_Country'] +" - "+played['Club_Liga1']
```

```python
played['League_ID'].value_counts()
```

## How many different leagues are represented among the players of the 2014 WorldCup?

```python
len(played['League_ID'].value_counts())
```

## How many different leagues are represented in each national team?

```python
played.groupby(by='Nationalmannschaft')['League_ID'].nunique().sort_values(ascending = True)
```

## (a) How many players of each league play in each national team?

```python
played.groupby(by='Nationalmannschaft')['League_ID'].value_counts()
```

## (b) How many players are in each national team?

```python
played.groupby(by='Nationalmannschaft')['Spieler'].nunique()
```

## (c) How many players has each league sent to a World Cup team?

```python
played['League_ID'].value_counts()
```
